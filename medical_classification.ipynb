```python
e['smoker'] = np.where(e['smoker'] == 'yes', 1, e['smoker'])
e['smoker'] = np.where(e['smoker'] == 'no', 0, e['smoker'])
print(e)
```

```python
#R ^ 2.
import statsmodels.api as sm
X = e["age"]
y = e["charges"]
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
predictions = model.predict(X)
print(model.summary())

X = e["bmi"]
y = e["charges"]
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
predictions = model.predict(X)
print(model.summary())
```
